```python
import json
```

```python
# to load the data
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)

# to find people you may know
def people_you_may_know(user_id,data):
    user_friends = {}   # creating an empty dict to store all the friends of the user.
    for user in data["users"]:
        user_friends[user['id']] = set(user['friends']) # storing direct friends
    if user_id not in user_friends:   # to remove invalid id
       return []
    direct_friends =user_friends[user_id] # storing the direct_friend in list
    suggestion = {}  # storing suggested friend(key) and their count(value)
    for i in direct_friends:     
        for j in user_friends[i]:
            if j!=user_id and j not in direct_friends: # to not repeat the same friend again and to not take the direct friends.
                # count the suggestions
                suggestion[j] = suggestion.get(j,0)+1
    sorted_suggestion = sorted(suggestion.items(), key = lambda x:x[1] , reverse=True)     
    return [user_id for user_id, _ in sorted_suggestion] # to return the suggested friends only.

# load the data
data = load_data("Data_of_coders.json")
data
user_id = 9
recc = people_you_may_know(user_id,data)
print(recc)
    
```
