```python
import json
```

```python
# load the file
def load_data(filename):
    with open(filename,"r") as f:
        return (json.load(f))

# function to find pages you might like
def pages_might_like(user_id,data):
    user_pages = {}    # create a ditionary to store pages that the user liked.
    # populate the dictionary
    for user in data["users"]:  
        user_pages[user['id']] = set(user['liked_pages'])
    if user_id not in user_pages:
        return []
    user_liked_pages = user_pages[user_id]  
    page_suggestion = {}  # created a dictionary to store page_id and page score. 
    for other_user,page in user_pages.items():  # to store others like pages as our suggestions
        if other_user!=user_id:  # so that it doesnot compare it with itself
            shared_page = user_liked_pages.intersection(page)   # finding the common pages between current user and other user.
            for i in page:
                if i not in user_liked_pages:
                    page_suggestion[i]= page_suggestion.get(i,0) + len(shared_page)
    sorted_pages = sorted(page_suggestion.items(), key = lambda x: x[1] , reverse= True)
    return [page_id for page_id, _ in sorted_pages]
        
data = load_data("Data_of_coders.json")
data
rec = pages_might_like(4,data)
print(rec)
        
```
