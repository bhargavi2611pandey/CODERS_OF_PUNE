```python
import json
# Function to load the json file in python.
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)   
    
```

```python
data = load_data("Data_of_coders.json")
data
```

TASK - 2 NOW SUPPOSE THE DATA HAS SOME MISSING VALUES AND DUPLICATE VALUES , SO WE HAVE TO CLEAN THE DATA.(data2.json)

```python
# load the data
import json
# we will use the above function which we have created .
data3 = load_data("Data_of_coders.json")
data3
```

```python
# function to clean the data
def clean_data(data):
    # remove the missing values
    data["users"]= [user for user in data["users"] if user['name'].strip() ] # used list comprehension to remove the missing value
    
    # remove the duplicate values
    for user in data["users"]:
        user['friends']=list(set(user['friends'])) # we want a list but also set so we have used both
        
    #remove inactive users
    data["users"] = [user for user in data["users"] if user['friends'] or user['liked_pages']]

    # remove duplicate page id (given: we have to store only one id with one page name)
    unique_pages = {}   # empty dict to store the unique id value 
    for page in data["pages"]:
        unique_pages[page['id']] = page  # 'id' in dictionary page has become key and now it won't repeat .
    data["pages"] = list(unique_pages.values())    # now updating the original dictionary with unique_pages values.
    
    return (data)

data3 = load_data("Data_of_coders.json")
data3    
    

data_c=clean_data(data3) # running the function for the current data
print(data_c)

json.dump(data_c,open("Data_of_coders.json","w"),indent=2 ) # now dump the python file back to json file 
print("Data has been cleaned successfully")
```
